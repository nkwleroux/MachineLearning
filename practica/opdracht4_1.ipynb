```python
# imports np from numpy.
import numpy as np
# used to display arrays and other things such as graphs inline/ inbetween the code.
%matplotlib inline
# imports plt from matplotlib.pyplot
import matplotlib.pyplot as plt
```

```python
# imports make_circles from sklearn.datasets. 
from sklearn.datasets import make_circles
```

```python
# assigns output from make_circles to variables X and y.
# X = The generated samples.
# y = The integer labels (0 or 1) for class membership of each sample. 
X, y = make_circles(n_samples=1000,
                    noise=0.1,
                    factor=0.2,
                    random_state=0)
```

```python
# displays the array/output from make_circles.
X
```

```python
# tupple with value 1 (1000) being the n amount of samples and value 2 (2) a default value.
X.shape
```

```python
# draws the size of the graph.
plt.figure(figsize=(5, 5))
# plots ob (blue circles).
plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
# plots xr (red crosses).
plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)
# sets the limit of the x axis.
plt.xlim(-1.5, 1.5)
# sets the limit of the y axis.
plt.ylim(-1.5, 1.5)
# adds a legend depicting what ob and xr are.
plt.legend(['0', '1'])
# adds a title to the graph.
plt.title("Blue circles and Red crosses")
```

```python
# imports Sequential from tensorflow.keras.models
from tensorflow.keras.models import Sequential
# imports Dense from tensorflow.keras.layers 
from tensorflow.keras.layers import Dense
# imports SGD from tensorflow.keras.optimizers.
from tensorflow.keras.optimizers import SGD
```

```python
# initiates model.
model = Sequential()
```

```python
# adds a Dense layer as the first layer to model with 4 units, (None, 2), and an activation of tanh. 
model.add(Dense(4, input_shape=(2,), activation='tanh'))
```

```python
# adds a Dense layer with an activation of sigmoid.
model.add(Dense(1, activation='sigmoid'))
```

```python
# configures the model for training. Uses a Gradient descent (with momentum) optimizer with a learning rate of 0.5.
model.compile(SGD(learning_rate=0.5), 'binary_crossentropy', metrics=['accuracy'])
```

```python
# trains the model for a fixed number of epochs (iterations on a dataset).
model.fit(X, y, epochs=20)
```

```python
# generates evenly-spaced values in an interval along a given axis.
hticks = np.linspace(-1.5, 1.5, 101)
# generates evenly-spaced values in an interval along a given axis.
vticks = np.linspace(-1.5, 1.5, 101)
# creates a meshgrid with the 2 values.
aa, bb = np.meshgrid(hticks, vticks)
ab = np.c_[aa.ravel(), bb.ravel()]
# generates output predictions for the input samples.
c = model.predict(ab)
# reshapes the shape according to the prediction.
cc = c.reshape(aa.shape)
```

```python
# draws the size of the graph.
plt.figure(figsize=(5, 5))
# draws a contour around the given predicted area.
plt.contourf(aa, bb, cc, cmap='bwr', alpha=0.2)
# plots ob (blue circles).
plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
# plots xr (red crosses).
plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)
# sets the limit of the x axis.
plt.xlim(-1.5, 1.5)
# sets the limit of the y axis.
plt.ylim(-1.5, 1.5)
# adds a legend depicting what ob and xr are.
plt.legend(['0', '1'])
# adds a title to the graph.
plt.title("Blue circles and Red crosses")
```
