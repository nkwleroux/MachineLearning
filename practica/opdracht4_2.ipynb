```python
# Merges the columns 'age' and 'chol'
hrt1 = hrt[['age','chol']]

hrt1Female = hrt[hrt['sex'] < 1]
hrt1Male = hrt[hrt['sex'] > 0]
```

```python
# Makes variable fig
fig=plt.figure()
# Adds subplot 1
ax1=fig.add_subplot(2,2,1)
# Adds subplot 2
ax2=fig.add_subplot(2,2,2)

# Plots female graph
hrt1Female.plot(ax=ax1,kind='scatter', x='age', y='chol',
            xlim=(30, 80), ylim=(100, 600), subplots= True)

# Plots male graph
hrt1Male.plot(ax=ax2,kind='scatter', x='age', y='chol',
            xlim=(30, 80), ylim=(100, 600), subplots= True)

# Draws all graphs in tight layout
plt.tight_layout()
```

```python
# draws the size of the graph.
plt.figure(figsize=(5, 5))
# plots female.
plt.plot(hrt1Female['age'], hrt1Female['chol'], 'ob', 0.5)
# plots male.
plt.plot(hrt1Male['age'], hrt1Male['chol'], 'xr', 0.5)
# Sets the axis of the graph
plt.axis([25, 80, 100, 600])
# adds a legend.
plt.legend(['Female', 'Male'])
# adds a title to the graph.
plt.title("Chol of both male and females by age")
```

```python
#Plots a graph with all data of the type historgram
hrt.plot(kind='hist',
        bins=50,
        title='Histogram',
        alpha=0.6);
```

```python
# Inits piecounts with 'sex' values total
piecounts = hrt['sex'].value_counts()

# Plots pie chart
piecounts.plot(kind='pie',
               figsize=(5, 5),
               explode=[0, 0.15],
               labels=['Male', 'Female'],
               autopct='%1.1f%%',
               shadow=True,
               startangle=90,
               fontsize=16);
```

```python
#define variable for height and weight
height = df['height']
weight = df['weight']
```

```python
#transpose index and columns
data = np.vstack([height, weight]).transpose()
```

```python
#show height and weight of the first 30 players in a table
df1 = pd.DataFrame(data, columns=['height', 'weight'])
df1.head(30)
```

```python
#create a line plot of the heights and weights
df1.plot(title='Line plot height and weight')
```

```python
#create a scatter plot of the heights and weights
df1.plot(style='.')
```

```python
# create a histogram of present heights and weights
df1.plot(kind='hist', #type = histogram
        bins=100, #100 pipes (for accuracy)
        title='Histogram of heights and weights',
        alpha=1.0)
```

```python
#create a boxplot for the heights and weights
df1.plot(kind='box',
        title='Boxplot of heights and weights')
```

```python
#show the just worked out plots in one go
fig, ax = plt.subplots(2, 2, figsize=(10, 10)) #create a 2x2 subplot

df1.plot(ax=ax[0][0], #line plot upper left
        title='Line plot') #title of plot

df1.plot(ax=ax[0][1], #scatter plot upper right
        style='o', #dot style
        title='Scatter plot') #title of plot

df1.plot(ax=ax[1][0], #histogram bottom left
        kind='hist', #histogram type
        bins=75, #75 bins
        title='Histogram') #title of plot

df1.plot(ax=ax[1][1], #boxplot bottom right
        kind='box', #boxplot type
        title='Boxplot') #title of plot

plt.tight_layout()
```
